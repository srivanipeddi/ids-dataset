```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ids=pd.read_csv('IDSCountry.csv')
```

```python
ids
```

```python
#hideeN columns
pd.set_option("display.max_columns",None)
```

```python
ids[ids.duplicated()]
```

```python
ids.drop_duplicates()
```

```python
ids.isnull().sum()
```

```python
#drop unwated columns
ids.drop(['Table Name','Long Name','Currency Unit','WB-2 code','National accounts reference year','Other groups','Alternative conversion factor','PPP survey year','Vital registration complete','Latest agricultural census','Latest industrial data'],axis=1,inplace=True)
```

```python
#fill nullvalues
data1=['Special Notes','Region','Income Group','Latest household survey','National accounts base year','SNA price valuation','Lending category','System of National Accounts','Balance of Payments Manual in use','External debt Reporting status','System of trade','Government Accounting concept','IMF data dissemination standard','Latest population census','Source of most recent Income and expenditure data','Latest trade data','Latest water withdrawal data']

for x in data1:
    ids[x]=ids[x].fillna(method='ffill')
```
